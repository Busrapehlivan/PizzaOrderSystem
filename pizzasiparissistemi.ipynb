```python
import csv 
import datetime 
import time
#menu bir txt dosyasına yazdırılır.
with open("Menu.txt", "w") as f:
    
    f.write("Lütfen pizzanızı seçiniz.:\n")
    f.write("1: Klasik 90.75₺\n")
    f.write("2: Margarita 98.75₺\n")
    f.write("3: TurkPizza 97.99₺\n")
    f.write("4: SadePizza 80.25₺\n")
    f.write("\n")
    f.write("ve Sos seçimleriniz:\n")
    f.write("11: Zeytin 10.99₺\n")
    f.write("12: Mantar 15.99₺\n")
    f.write("13: KeçiPeyniri 18.99₺\n")
    f.write("14: Et 30.99₺\n")
    f.write("15: Soğan 11.99₺\n")
    f.write("16: Mısır 12.99₺\n")
    f.write("\n")
    f.write("Teşekkür Ederiz!")
#pizza sınıfı oluşturulur.
#init metodu ile parametreler tanımlanır.

class Pizza: 
    
    def __init__(self, description='Pizza', cost=0):
        self.description = description 
        self.cost = cost 
    
    def get_description(self): 
        return self.description 

    def get_cost(self): 
        return self.cost

#Pizza alt sınıfları oluşturulur ve fiyat tanımlamaları yapılır.

class KlasikPizza(Pizza): 
    def __init__(self):
        super().__init__('Klasik Pizza', 90.75) 

class MargaritaPizza(Pizza):
    def __init__(self):
        super().__init__('Margarita Pizza', 98.75)

class TurkPizza(Pizza):
    def __init__(self):
        super().__init__('Turk Pizza', 97.99)

class SadePizza(Pizza):
    def __init__(self):
        super().__init__('Sade Pizza', 80.25)

#dekarator super sınıfları oluşturulur.

class Decorator(Pizza): 
#dekoratör sınıfı, ekstra malzemeleri eklemek için kullanılır.
    def __init__(self, component, description, cost): 
      #__init__ metodu, dekoratör sınıfının bileşenini, açıklamasını ve maliyetini tanımlar. 
      #get_cost metodu, dekoratör bileşeninin maliyetini, Pizza sınıfından gelen maliyet ile toplayarak hesaplar.
      #get_description metodu, dekoratör bileşeninin açıklamasını, Pizza sınıfından gelen açıklama ile birleştirerek döndürür.
        self.component = component
        self.description = description
        self.cost = cost 
    
    def get_cost(self): #Pizza sınıfından gelen maliyet ile toplar ve toplam maliyeti döndürür.
        return self.component.get_cost() + Pizza.get_cost(self)

    def get_description(self): #get_description metodu, dekoratör bileşeninin açıklamasını, Pizza sınıfından gelen açıklama ile birleştirerek döndürür.
        return self.component.get_description() + ' ' + Pizza.get_description(self)

#dekarator alt sınıflarının fiyatları tanımlanır.

class Zeytin(Decorator):
    def __init__(self, component):
        super().__init__(component, 'Zeytin, ', 10.99)

class Mantar(Decorator):
    def __init__(self, component):
        super().__init__(component, 'Mantar, ', 15.99)

class KeciPeyniri(Decorator):
    def __init__(self, component):
        super().__init__(component, 'Keci Peyniri, ', 18.99)

class Et(Decorator):
    def __init__(self, component):
        super().__init__(component, 'Et, ', 30.99)

class Sogan(Decorator):
    def __init__(self, component):
        super().__init__(component, 'Soğan, ', 11.99)

class Misir(Decorator):
    def __init__(self, component):
        super().__init__(component, 'Mısır, ', 12.99)

#csv kütüphanesinden faydalanılır.
       
def write_to_csv(isim, tc, kk, son_hesap, odeme_tarih, kart_psw):
    
  try:
      with open("Orders_Database.csv") as csv_file:
        pass

  except FileNotFoundError:
      with open("Orders_Database.csv", "w") as csv_file:
        fieldnames = ['isim', 'tc', 'kk', "son_hesap", "odeme_tarih", "kart_psw"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()

      with open("Orders_Database.csv", "a", newline='') as csv_file:
        fieldnames = ['isim', 'tc', 'kk', "son_hesap", "odeme_tarih", "kart_psw"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        writer.writerow({
          "isim": isim,
          "tc": tc,
          "kk": kk,
          "son_hesap": son_hesap,
          "odeme_tarih": datetime.datetime.now().strftime('%Y-%m-%d %H:%M'),
          "kart_psw": kart_psw
    })

def main(): #main() fonksiyonu, kesintiye kadar devam edecek bir while döngüsü içerir.
    
    while True:
        
        print('Hosgeldiniz!\n\nMenumuz:\n')
        
        print(open('Menu.txt', 'r').read())
                    
#kullanıcının 1 ile 4 arasında bir sayı girerek bir pizza seçmesini isteyen başka bir while döngüsüne girer.      
        while True:

#Kullanıcı geçersiz bir giriş yaparsa (yani, bir sayı değil veya 1 ila 4 arasında değil), 
#hata mesajı verilir ve kullanıcıdan doğru bir giriş yapması istenir.           
            try:
                
                pizza_sec = int(input("Lütfen bir pizza seçiniz (1-4): "))
                
                if pizza_sec < 1 or pizza_sec > 4:
                    print("Lütfen geçerli bir seçim yapınız.")
                
                else:
                    break
            
            except ValueError:
                print("Lütfen geçerli bir sayı giriniz.")

        # Sos seçimi için boş liste açılır.
        
        sos_list=[]
        
        while True: #while döngüsü, kullanıcıya sos seçenekleri hakkında sorular sorar ve doğru bir giriş yapana kadar devam eder.
            try:
                
                sos_sec = int(input("Lütfen bir sos seçiniz (11-16)\n(Ödemeye geçmek için '0' basınız.):"))
                #Eğer kullanıcı 0 tuşuna basarsa, bu döngüyü kırar ve sos seçim işlemi sona erer.
               
                if sos_sec==0:
                    break
                
                elif sos_sec < 11 or sos_sec > 16:
                    print("Lütfen geçerli bir seçim yapınız.")
               
                else:
                    sos_list.append(sos_sec)
                    continue
           
            except ValueError:
                
                print("Lütfen geçerli bir sayı giriniz.")
        
        print(sos_list)
       
        while sos_sec==0:
        # kullanıcının seçtiği pizzaya göre bir pizza nesnesi oluşturur.
           
            if pizza_sec == 1:
                pizza = KlasikPizza()
                break
            elif pizza_sec == 2:
                pizza = MargaritaPizza()
                break
            elif pizza_sec == 3:
                pizza = TurkPizza()
                break
            else:
                pizza = SadePizza()
                break

        
        pizza_base=pizza.get_cost()
        
        sos_cost=0
        
        sos_description=""
        
        for element in sos_list:
                
                if element == 11:
                    sos_cost += Zeytin(pizza).cost
                    sos_description+=" "+Zeytin(pizza).description
                
                elif element == 12:
                    sos_cost += Mantar(pizza).cost
                    sos_description+=" "+Mantar(pizza).description
                    
                elif element == 13:
                    sos_cost += KeciPeyniri(pizza).cost
                    sos_description+=" "+KeciPeyniri(pizza).description

                elif element == 14:
                    sos_cost += Et(pizza).cost
                    sos_description+=" "+Et(pizza).description

                elif element == 15:
                    sos_cost += Sogan(pizza).cost
                    sos_description+=" "+Sogan(pizza).description

                else:
                    sos_description+=" "+Misir(pizza).description
                    sos_cost +=Misir(pizza).cost
        
        # Toplam fiyat ve açıklama, pizzanın açıklamasına ve seçilen sosların açıklamalarına dayanarak hesaplanır.
        total_cost = pizza_base+sos_cost
        description = pizza.get_description()+sos_description
        
        print('\nSiparisiz Ozetiniz: \nPizza seçiminiz: {}\nSos seçiminiz: {}\nTutar: {} ₺ '.format(pizza.get_description(), sos_description, total_cost))


        
        
        isim = input("Lütfen adınızı giriniz: ")
        
        tc = input("Lütfen TC numaranızı giriniz: ")
               
        kk = input("Kredi Kartı numaraınızı giriniz: ")
                
        kart_psw = input("Kredi kartı şifrenizi giriniz: ")
                
            
        print("\nÖdemeniz Başarılı!")
        
        #ödeme tarihi kayıt edilir.
        odeme_tarih = datetime.datetime.now()
        
        write_to_csv(isim, tc, str(kk), total_cost, odeme_tarih, str(kart_psw))

        
        yeni_sparis = input("Yeni spariş oluşturmak istiyor musunuz? Evet(1)/Hayır(0)?: ")
        
        if yeni_sparis== "1":
            continue
        
        elif yeni_sparis == "0":
                
            break

    # csv_yaz fonksiyonunun çağrılarak verilen bilgilerin database'e işlen
    # datetime modülünden şu anki tarih/saat bilgisini alır. 
    #İkinci satır, tarih/saat bilgisini istenen formata (YYYY-AA-GG SS:DD) dönüştürür. 
    now = datetime.datetime.now()
    formatted_time = now.strftime('%Y-%m-%d %H:%M') 
    print("Tebrikler siparisiniz basari ile olusturuldu.", formatted_time)

if __name__ == "__main__":
        main()
        print("İyi Günler Dileriz!") 


                       
        print("Afiyet Olsun...")
```
